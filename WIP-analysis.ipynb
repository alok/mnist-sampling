```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_out = np.load('test_losses.npy')
```

```python
losses = raw_out[:, 0].reshape((101, 10))
accuracies = raw_out[:, 1].reshape((101, 10))
```

## Losses

```python
# plot losses first
z = losses[1:, :]
x, y = np.mgrid[slice(1, 101, 1),
                slice(1, 11, 1)]
```

```python
levels = matplotlib.ticker.MaxNLocator(nbins=15).tick_values(z.min(), z.max())
cmap = plt.get_cmap('PiYG')
norm = matplotlib.colors.BoundaryNorm(levels, ncolors=cmap.N, clip=True)
```

```python
fig, ax = plt.subplots(nrows=1)
im = ax.pcolormesh(x, y, z, cmap=cmap)
fig.colorbar(im, ax=ax)
ax.set_title('Losses v. (percentage of overall dataset, epoch trained)')
```

```python
fig, ax = plt.subplots(nrows=1)
cf = ax.contourf(x, y, z, levels=levels, cmap=cmap)
fig.colorbar(cf, ax=ax)
ax.set_title('Losses as a func of (% of overall dataset, epochs trained)')
```
